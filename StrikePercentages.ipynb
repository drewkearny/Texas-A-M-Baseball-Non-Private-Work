```python
import pandas as pd
df = pd.read_csv('20240216-Olsen-1_unverified.csv')
```

```python
df['Count'] = df['Balls'].astype(str) + '-' + df['Strikes'].astype(str)
```

```python

# Filter for all plays with a 0-0 count
df_0_0 = df[(df['Balls'] == 0) & (df['Strikes'] == 0)]

# Further filter for PitcherTeam = 'TEX_AGG'
df_tex_agg = df_0_0[df_0_0['PitcherTeam'] == 'TEX_AGG']
df_oppo = df_0_0[df_0_0['PitcherTeam'] == 'MCN_COW']
# Define lists of PitchCall values for first-pitch strike and ball
first_pitch_strike_calls = ['StrikeCalled', 'InPlay', 'StrikeSwinging', 'FoulBallNotFieldable', 'FoulBallFieldable']
first_pitch_ball_calls = ['BallCalled', 'BallinDirt', 'HitByPitch']

# Filter for first-pitch strikes and balls based on PitchCall
first_pitch_strikes = df_tex_agg[df_tex_agg['PitchCall'].isin(first_pitch_strike_calls)]
first_pitch_balls = df_tex_agg[df_tex_agg['PitchCall'].isin(first_pitch_ball_calls)]
afirst_pitch_strikes = df_oppo[df_oppo['PitchCall'].isin(first_pitch_strike_calls)]
afirst_pitch_balls = df_oppo[df_oppo['PitchCall'].isin(first_pitch_ball_calls)]

# Calculate the count of each
num_first_pitch_strikes = first_pitch_strikes.shape[0]
num_first_pitch_balls = first_pitch_balls.shape[0]
anum_first_pitch_strikes = afirst_pitch_strikes.shape[0]
anum_first_pitch_balls = afirst_pitch_balls.shape[0]

# Calculate the total number of first pitches considered
total_first_pitches = num_first_pitch_strikes + num_first_pitch_balls
atotal_first_pitches = anum_first_pitch_strikes + anum_first_pitch_balls

# Calculate the percentage of each
first_pitch_strike_percentage = round((num_first_pitch_strikes / total_first_pitches) * 100,2) if total_first_pitches > 0 else 0
first_pitch_ball_percentage = round((num_first_pitch_balls / total_first_pitches) * 100,2) if total_first_pitches > 0 else 0
awayfirst_pitch_strike_percentage = round((anum_first_pitch_strikes / atotal_first_pitches) * 100,2) if atotal_first_pitches > 0 else 0
awayfirst_pitch_ball_percentage = round((anum_first_pitch_balls / atotal_first_pitches) * 100,2) if atotal_first_pitches > 0 else 0

# Print the results
print(f" Aggies First Pitch Strike Percentage: {first_pitch_strike_percentage}%")
print(f" Aggies First Pitch Ball Percentage: {first_pitch_ball_percentage}%")
print(f" Away First Pitch Strike Percentage: {awayfirst_pitch_strike_percentage}%")
print(f" Away First Pitch Ball Percentage: {awayfirst_pitch_ball_percentage}%")
```
